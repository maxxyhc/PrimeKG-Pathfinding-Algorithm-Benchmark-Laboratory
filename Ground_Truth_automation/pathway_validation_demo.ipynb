```python
validator = PathwayValidator('kg.csv')

print("\nFind all the pathwways")
paths = validator.discover_actual_pathway(
    drug='Finasteride',
    disease='androgenetic alopecia',
    max_depth=5
)

print(f"Find {len(paths)} pathsï¼š")
for i, path in enumerate(paths[:5], 1):  # show only first 5 paths
    print(f"  {i}. {' â†’ '.join(path)}")
```

```python


# Initialize validator
validator = PathwayValidator('kg.csv')

print("\n" + "="*70)
print("VALIDATING 5 DRUG REPURPOSING PATHWAYS")
print("="*70 + "\n")

# ========================================================================
# 1. Bupropion â†’ Nicotine Dependence
# ========================================================================
print("\n" + "="*70)
print("[1/5] Bupropion â†’ Nicotine Dependence")
print("="*70)

result1 = validator.validate_pathway(
    drug='Bupropion',
    disease='nicotine dependence',
    expected_pathway=[
        'Bupropion',
        'SLC6A3',
        'DRD2',
        'nicotine dependence'
    ]
)

print(f"\nPathway found in KG: {result1.found_in_kg}")
print(f"Coverage score: {result1.coverage_score:.1%}")

# ========================================================================
# 2. Minoxidil â†’ Androgenetic Alopecia
# ========================================================================
print("\n" + "="*70)
print("[2/5] Minoxidil â†’ Androgenetic Alopecia")
print("="*70)

result2 = validator.validate_pathway(
    drug='Minoxidil',
    disease='androgenetic alopecia',
    expected_pathway=[
        'Minoxidil',
        'REN',
        'kidney',
        'AR',
        'androgenetic alopecia'
    ]
)

print(f"\nPathway found in KG: {result2.found_in_kg}")
print(f"Coverage score: {result2.coverage_score:.1%}")

# ========================================================================
# 3. Sildenafil â†’ Erectile Dysfunction
# ========================================================================
print("\n" + "="*70)
print("[3/5] Sildenafil â†’ Erectile Dysfunction")
print("="*70)

result3 = validator.validate_pathway(
    drug='Sildenafil',
    disease='erectile dysfunction (disease)',
    expected_pathway=[
        'Sildenafil',
        'PDE5A',
        'cGMP effects',
        'Nitric oxide stimulates guanylate cyclase',
        'NOS3',
        'erectile dysfunction (disease)'
    ]
)

print(f"\nPathway found in KG: {result3.found_in_kg}")
print(f"Coverage score: {result3.coverage_score:.1%}")

# ========================================================================
# 4. Finasteride â†’ Androgenetic Alopecia
# ========================================================================
print("\n" + "="*70)
print("[4/5] Finasteride â†’ Androgenetic Alopecia")
print("="*70)

result4 = validator.validate_pathway(
    drug='Finasteride',
    disease='androgenetic alopecia',
    expected_pathway=[
        'Finasteride',
        'SRD5A2',
        'androgenetic alopecia'
    ]
)

print(f"\nPathway found in KG: {result4.found_in_kg}")
print(f"Coverage score: {result4.coverage_score:.1%}")

# ========================================================================
# 5. Finasteride â†’ Benign Prostatic Hyperplasia
# ========================================================================
print("\n" + "="*70)
print("[5/5] Finasteride â†’ Benign Prostatic Hyperplasia")
print("="*70)

result5 = validator.validate_pathway(
    drug='Finasteride',
    disease='benign prostatic hyperplasia',
    expected_pathway=[
        'Finasteride',
        'SRD5A2',
        'benign prostatic hyperplasia'
    ]
)

print(f"\nPathway found in KG: {result5.found_in_kg}")
print(f"Coverage score: {result5.coverage_score:.1%}")

# ========================================================================
# FINAL SUMMARY
# ========================================================================
print("\n" + "="*70)
print("FINAL SUMMARY")
print("="*70)

pathways = [
    ("Bupropion â†’ Nicotine Dependence", result1),
    ("Minoxidil â†’ Androgenetic Alopecia", result2),
    ("Sildenafil â†’ Erectile Dysfunction", result3),
    ("Finasteride â†’ Androgenetic Alopecia", result4),
    ("Finasteride â†’ Benign Prostatic Hyperplasia", result5)
]

valid_count = 0
for name, result in pathways:
    status = "âœ“" if result.found_in_kg else "âœ—"
    if result.found_in_kg:
        valid_count += 1
    print(f"{status} {name:45} | Coverage: {result.coverage_score:>6.1%} | Found: {result.found_in_kg}")

print(f"\n{'='*70}")
print(f"Valid pathways: {valid_count}/5 ({valid_count/5:.1%})")
print(f"{'='*70}\n")
```
